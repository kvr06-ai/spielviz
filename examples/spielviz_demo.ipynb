## 1. Quick Render

```python
import spielviz

fig, ax = spielviz.quick_render("kuhn_poker")
```

## 2. Game Tree with Information Sets

```python
import pyspiel

game = pyspiel.load_game("kuhn_poker")
root = spielviz.build_tree(game)
root = spielviz.layout_tree(root)

colors = spielviz.auto_color_info_sets(root, player=0)
fig, ax = spielviz.render_tree(
    root,
    highlight_info_sets=colors,
    show_actions=True,
    show_payoffs=True,
    title="Kuhn Poker — Player 0 Information Sets",
)
```

## 3. Information Set Connections

```python
fig, ax = spielviz.render_tree(root, highlight_info_sets=colors)
spielviz.draw_info_set_connections(ax, root, player=0)
fig
```

## 4. CFR Strategy Profile

```python
from open_spiel.python.algorithms import cfr

solver = cfr.CFRSolver(game)
for _ in range(1000):
    solver.evaluate_and_update_policy()

policy = solver.average_policy()
fig, ax = spielviz.plot_strategy_profile(
    game, policy, player=0,
    title="Kuhn Poker — CFR Equilibrium (Player 0)",
)
```

## 5. Strategy on Tree

```python
fig, ax = spielviz.plot_strategy_on_tree(root, game, policy, player=0)
```
